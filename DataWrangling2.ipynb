```python
from lxml import html
import requests
import html5lib
import BeautifulSoup
import urllib2
import sys
import csv
from BeautifulSoup import BeautifulSoup
import pandas as pd
```

```python
cik='0000051143'
acc_no='0000051143-13-000007'
acc_no_short='000005114313000007'
page=requests.get('https://www.sec.gov/Archives/edgar/data/'+cik+'/'+acc_no_short+'/'+acc_no+'-index.html')
source=page.content
htmlcontent=html.document_fromstring(source)
href=''
for elem in htmlcontent.iterlinks():
    if "10q.htm" in elem[2]:
        href+=elem[2]
url10q='https://www.sec.gov'+href
```
